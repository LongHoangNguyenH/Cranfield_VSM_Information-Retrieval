```python
def remove_special_character(txt):
    text_processed = ''
    for char in txt:
        if char.isalpha() or char==' ':
            text_processed+=char
    return text_processed
```

## TASK 1:  Statistics of words used in Cranfield documents to know how many entries, how many times each item appears and how many documents appear.

```python
doc = dict()
Cranfield = []
for file in file_name:
    text = open(dir_path+file,'r')
    for line in text:
        line = line.strip()
        line = remove_special_character(line.lower())
        words = line.split()
    for word in words:		      
        if word in doc:
            doc[word] = doc[word] + 1
        else:			
            doc[word] = 1
    #Bước này có thể cải tiến sau
    with open(dir_path+file,'r') as f:
        lines = f.readlines()
        lines = ''.join(lines).strip()
        remove_special_character(lines.lower())
        #Cranfield.append(lines)
        temp=file.replace('L:/MonHocDaiHoc/CS419_TruyXuatThongTin/BaiTapNhom1/Cranfield/','')
        Cranfield.append([lines,temp.replace('.txt','')])
        
```

##TASK 2: Define Term in Cranfield Dataset

Term là những từ 
+ Separate by whitespace
+ Not in stopword list
+ Has been stemming

```python
from nltk.corpus import stopwords
import re
import glob
from nltk.tokenize import sent_tokenize , word_tokenize
from nltk.stem import PorterStemmer
import nltk
```

```python

nltk.download('stopwords')
Stopwords = set(stopwords.words('english'))
```

```python
stopwords
```

```python
# these function maybe repeat the code above in Task1 - you can upgrade it, if you want :))
def finding_all_unique_words_and_freq(words):
    words_unique = []
    word_freq = {}
    for word in words:
        if word not in words_unique:
            words_unique.append(word)
    for word in words_unique:
        word_freq[word] = words.count(word)
    return word_freq
def finding_freq_of_word_in_doc(word,words):
    freq = words.count(word)
        
def remove_unvalid_characters(text):
    regex = re.compile('[^a-zA-Z0-9\s]')
    text_returned = re.sub(regex,'',text)
    return text_returned
```

```python
all_words = []
dict_global = {}
file_folder = "/content/drive/MyDrive/BaiTapNhom1_CS419/Cranfield/*"
idx = 1
files_with_index = {}
nltk.download('punkt')
for file in glob.glob(file_folder):
    # print(file)
    fname = file
    file = open(file , "r")
    text = file.read()
    text = remove_unvalid_characters(text)
    text = re.sub(re.compile('\d'),'',text)    
    sentences = sent_tokenize(text)
    words = word_tokenize(text)
    words = [word for word in words if len(words)>1]
    words = [word.lower() for word in words]
    words = [word for word in words if word not in Stopwords]
    dict_global.update(finding_all_unique_words_and_freq(words))
    files_with_index[idx] = os.path.basename(fname)
    idx = idx + 1
    
unique_words = set(dict_global.keys())
```

```python
st = PorterStemmer()
unique_words_process = {}
for item in unique_words:
    unique_words_process.update({st.stem(item):''})
unique_words_processed = set(unique_words_process.keys())
```
